```python
import seaborn as sns

import numpy as np
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')  #loading a built in dataset
```

```python
tips.head()
```

```python
sns.distplot(tips['total_bill'])  #generates a distribution plot
```

```python
sns.distplot(tips['total_bill'], kde = False)  #dist plot without kde
```

```python
sns.distplot(tips['total_bill'], kde = False, bins = 40 )  #dist plot without kde and with bins
```

```python
sns.jointplot(x = 'total_bill' ,y = 'tip', data=tips)  #joint plots to compare two columns
```

```python
sns.jointplot(x = 'total_bill' ,y = 'tip', data=tips, kind = 'hex')  #kind of jointplot that we want to see, here we cahnged it from scatter to hex 
```

```python
sns.jointplot(x = 'total_bill' ,y = 'tip', data=tips, kind = 'reg') #plots a regression line instead of hex
```

```python
sns.jointplot(x = 'total_bill' ,y = 'tip', data=tips, kind = 'kde') # inserts a kde and shows the density instead of regression line
```

```python
sns.pairplot(tips) #quickly visualize the relationship between the columns of your data with pairplots
```

```python
sns.pairplot(tips, hue = 'sex')   #adding a hue will differentiate what we want to see, here we're comparing the data points b/w male and female customers
```

```python
sns.pairplot(tips, hue = 'sex', palette = 'coolwarm')  #palette adds a color to your plots
```

```python
sns.rugplot(tips['total_bill'])  #draws a line for each point on the bins of a distribution plot along the axis
```

```python
sns.displot(tips['total_bill'])
```

```python
sns.kdeplot(tips['total_bill'])  #kernel density estimation plot
            
```

```python
sns.barplot(x = 'sex', y = 'total_bill' ,data = tips)   #categorical plots using barplot, here it shows mean total bill for each of the categories
```

```python
sns.barplot(x = 'sex', y = 'total_bill' ,data = tips, estimator = np.std )   #shows the standard deviation, used np.std to to do this
```

```python
sns.countplot(x = 'sex', data = tips) #counting the occurences using countplot
```

```python
tips.describe()
```

```python
tips.tail()
```

```python
sns.boxplot(x = 'day' ,y = 'total_bill' , data = tips)  #using box plots
```

```python
sns.boxplot(x = 'day' ,y = 'total_bill' , data = tips, hue = 'smoker')
```

```python
sns.violinplot( x = 'day', y = 'total_bill', data = tips)  #using violin plots
```

```python
sns.violinplot( x = 'day', y = 'total_bill', data = tips, hue = 'sex')
```

```python
sns.violinplot( x = 'day', y = 'total_bill', data = tips, hue = 'sex', split = True)
```

```python
sns.stripplot(x = 'day', y = 'total_bill', data = tips) #strip plots
```

```python
sns.stripplot(x = 'day', y = 'total_bill', data = tips, jitter = True)
```

```python
sns.stripplot(x = 'day', y = 'total_bill', data = tips, jitter = True, hue = 'sex')
```

```python
sns.swarmplot(x = 'day', y = 'total_bill', data = tips, color = 'orange') #swarm plots
```

```python
tips.corr(numeric_only= True)  #finds correlation between numeric values
```

```python
tc = tips.corr(numeric_only= True)
```

```python
sns.heatmap(tc)  #generating a heat map
```

```python
sns.heatmap(tc, annot = True) #annoting with numbers in heatmap
```

```python
sns.heatmap(tc, annot = True, cmap = 'coolwarm')
```
